```python
from pyquibbler import iquib, quibbler_user_function
from pyquibbler.graphics.overriding import override_applier_functions, override_axes_methods
import matplotlib.pyplot as plt
```

```python
override_axes_methods()
override_applier_functions()
```

```python
my_quib = iquib([1, 2, 3, 4])
```

```python
@quibbler_user_function(lazy=True)
def plot_a_few_stuff(q):
    plt.plot(q)
```

```python
plot_quib = plot_a_few_stuff(my_quib)
```

```python
plot_quib.run_if_needed_and_draw()
```

```python
for i in range(20):
    my_quib[0] = i
```

```python
@quibbler_user_function(lazy=False)
def plot_a_few_stuff_eagerly(q):
    plt.plot(q)
```

```python
plot_a_few_stuff_eagerly(my_quib)
```
